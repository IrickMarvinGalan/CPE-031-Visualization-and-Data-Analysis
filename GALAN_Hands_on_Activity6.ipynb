```python
df2 = pd.read_excel('/content/drive/MyDrive/CPE_031_Galan/SampleData.xlsx')
df2
```

```python
df2 = df2[df2['Units'] > 30]
df2
```

```python
import seaborn as sns

# Pivot table: Cities (rows) vs. Outcomes (columns)
heat_data = df2.pivot_table(
    index="Region",
    columns="Item",
    values="Units",
    aggfunc="sum"
).fillna(0)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heat_data, cmap="Blues", linewidths=0.5)

plt.title("Regions Dominating School Supply Selling")
plt.xlabel("Item Type")
plt.ylabel("Regions")
plt.show()
```

Intepretation:
Through this heat map, we could visualize which regions dominate the selling of certain items procured from the main supplier. First, I filtered the individuals who purchased items greater than 30, because certainly 30 pieces are more than for personal use and summed their totals per region. After all that preparation, we could interpret that Central seems to dominate the market of supplies with the most purchases compared to other regions. West is the least active in procuring supplies, and East strikes a balance in its purchase quantity.

```python
df3 = pd.read_excel('/content/drive/MyDrive/CPE_031_Galan/SampleData.xlsx')
```

```python
df3 = df3[df3['OrderDate'] < '2022-01-01']
df3 = df3[['OrderDate', 'Total']]
df3
```

```python
# Group by City and sum counts
total_sales = df3.groupby('OrderDate')['Total'].sum()

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(total_sales.index, total_sales.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title('Total Sales per Date in 2021')
plt.xlabel('Date')
plt.ylabel('Sales Revenue')

# Rotate x-axis labels for readability
plt.xticks(rotation=0)

# Show chart
plt.show()
```

In this graph I filtered out the dates past 2022. I grouped the line chart by Order Date and Total Revenue per date. This allowed us to visualize what months and dates had the most sales in the annual year of 2021. We could see that January and August seem to be most active in sales, we may speculate that this is due to the patterns of resumption of classes, usually occuring in January after Winter Break, and during August after Summer Break. This proves to me that the data seem to make sense in the real world.

```python
df4 = pd.read_excel('/content/drive/MyDrive/CPE_031_Galan/SampleData.xlsx')
df4 = df4[df4['Rep'] == 'Jones']
df4
```

```python
# Group by City and sum counts
Jtotal_sales = df4.groupby('OrderDate')['Units'].sum()

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(Jtotal_sales.index, Jtotal_sales.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title('Jones\' Purchasing Pattern')
plt.xlabel('Date')
plt.ylabel('Number of Units Sold')

# Rotate x-axis labels for readability
plt.xticks(rotation=0)

# Show chart
plt.show()
```

In this graph, we inspect the purchasing pattern of one of our most frequent buyer named Jones. At initial purchases, he seemed to have procured greater quantities of items and seems to decline upredictably, but it never reached the same number of units as it originally was. This might mean that Jones had now predicted how much his market purchases after some time and avoids overpurchasing.
