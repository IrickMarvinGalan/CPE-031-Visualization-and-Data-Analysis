**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
# Highlight the worst performing month (preattentive color cue)
highlight = df["Sales"].idxmax()
lowest = df["Sales"].idxmin()

colors = ["gray" if i != highlight else "green" for i in range(len(df))]
colors = ["crimson" if i == lowest else colors[i] for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Emphasizing Month of Maximum Minimimum Sales", fontsize=13)
plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

The business is something that people seem to be patronized during the start of year but fell off in March. However, since starting April showed a promising recovery and even an increase, this tells us a story that the business might have something to do with summer events, or even school season.

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/CPE_031_Galan/diabetes.csv')

diab_df = pd.DataFrame(dataset)

diab_df = diab_df[["BloodPressure", "Age", "Outcome"]]

print(diab_df)
```

```python
# Filter for diabetic outcomes
diabetic_ages = diab_df[diab_df['Outcome'] == 1]['Age']

# Define age ranges
age_bins = range(20, 90, 10) # Bins from 20 to 80 with step 10

# Group ages into bins and count occurrences
age_range_counts = diabetic_ages.groupby(pd.cut(diabetic_ages, bins=age_bins, right=False)).size()

# Find the age range with the highest count of diabetic outcomes
highest_prob_range = age_range_counts.idxmax()

# Create labels for the age ranges
age_range_labels = [f'{interval.left}-{interval.right-1}' for interval in age_range_counts.index]

# Create a bar chart of age ranges for diabetic outcomes
plt.figure(figsize=(16, 8))

# Define colors for the bars, highlighting the highest probability age range
colors = ['skyblue' if interval != highest_prob_range else 'crimson' for interval in age_range_counts.index]

plt.bar(age_range_labels, age_range_counts.values, color=colors, edgecolor='black')

plt.title('Distribution of Age Ranges for Diabetic Outcomes', fontsize=14)
plt.xlabel('Age Range', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add annotation for the highest probability age range
plt.annotate(f'Highest occurrence: {highest_prob_range.left}-{highest_prob_range.right-1}',
             xy=(age_range_labels.index(f'{highest_prob_range.left}-{highest_prob_range.right-1}'), age_range_counts.max()),
             xytext=(age_range_labels.index(f'{highest_prob_range.left}-{highest_prob_range.right-1}') + 0.5, age_range_counts.max() + 5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()
```

The dataset shows that the age range of adolescents starts to see a rise in trend of diabetes outcomes. Unlike the previous years and decades, diabetes is often the disease of the old, but here we start to see a new trend of high diabetic probability in younger generations, probably due to diet nowadays.
