**3. Discussions:**

A well-designed data dashboard is more than a collection of charts — it’s a purposeful storytelling tool that connects data to decisions.

Understanding the Purpose of the Data Dashboard:
Every dashboard must have a clear objective — whether it’s to monitor performance, analyze patterns, or inform strategic decisions. The design should align with this goal, focusing only on the most relevant data and metrics.

Considering the Needs of the Data Dashboard’s Users:
Dashboards should be designed with the user in mind. A manager may need high-level KPIs, while an analyst may need granular, filterable data. Knowing the user’s goals, skills, and workflow ensures that the dashboard supports, rather than overwhelms, their decision-making.

Data Dashboard Engineering:
Engineering a dashboard involves organizing visual elements efficiently, ensuring smooth interactivity, and maintaining data accuracy. It also includes performance considerations (like handling large datasets), responsive design, and consistent color schemes.

Step 1: Install and Import Required Libraries

```python
!pip install plotly pandas
import pandas as pd
import numpy as np
import plotly.express as px
```

Step 2: Create a Sample Dataset

We’ll simulate student performance data to design an academic performance dashboard.

```python
# Create dataset
np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)
df.head()
```

Step 3: Design Dashboard Components

```python
# Average Score per Subject
fig1 = px.bar(df.groupby("Subject", as_index=False)["Score"].mean(),
              x="Subject", y="Score", title="Average Score per Subject",
              color="Score", color_continuous_scale="Blues")

# Attendance vs. Score Scatter
fig2 = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                  title="Attendance vs. Performance", hover_data=["Student"])

# Score Distribution
fig3 = px.box(df, x="Subject", y="Score", title="Score Distribution by Subject",
              color="Subject")

fig1.show()
fig2.show()
fig3.show()
```

Step 4: Combine into One Interactive Dashboard

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Aggregated data
avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

# Dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score",
                                    "Score Distribution", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

fig.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)
fig.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="Scatter Plot"), row=1, col=2)
fig.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig.update_layout(height=700, width=950, title_text="🎓 Student Performance Dashboard")
fig.show()
```

### **Teacher's Dashboard: Highlighting Underperforming Students**

For a teacher, the primary need is often to identify students who are struggling. This part of the dashboard focuses on visualizing and emphasizing students that fall below a certain score which could be passing grade or the professor's standard, allowing teachers to quickly spot those who struggle and might need help, advice, or remedial.

```python
underperform_threshold = 70

# Identify underperforming students (any score below a threshold in any subject)
underperforming_students_df = df[df['Score'] < underperform_threshold]

if not underperforming_students_df.empty:
    # Consolidate the scores by subject to see common weak areas of the selected student
    avg_underperform_score = underperforming_students_df.groupby(['Student', 'Subject'], as_index=False)['Score'].mean()

    fig_teacher = px.bar(avg_underperform_score,
                         x='Student', y='Score', color='Subject',
                         title=f'Underperforming Students (Scores < {underperform_threshold})',
                         barmode='group',
                         text='Score')
    fig_teacher.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig_teacher.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
    fig_teacher.show()
else:
    print(f"No students found with scores below {underperform_threshold}.")
```

### **Student's Dashboard: Personal Progress vs. Class Average**

This dashboard allows quick comparisons regards to what areas a student falls behind compared to what the class average does. I grouped the bars into clusters of two which allows quick visua comparison of a certain student score with the average. This also enables students to gauge their performance whether they all do struggle or peers could actually offer help in academics.

```python
sample_student = 'Student 5' # The current selected student in the dataset

student_scores = df[df['Student'] == sample_student].groupby('Subject', as_index=False)['Score'].mean()
student_scores.rename(columns={'Score': f'{sample_student} Score'}, inplace=True)

comparison_df = pd.merge(avg_scores, student_scores, on='Subject', how='left')
comparison_df.rename(columns={'Score': 'Class Average Score'}, inplace=True)

comparison_df[f'{sample_student} Score'].fillna(0, inplace=True)

fig_student = go.Figure(data=[
    go.Bar(name='Class Average', x=comparison_df['Subject'], y=comparison_df['Class Average Score']),
    go.Bar(name=sample_student, x=comparison_df['Subject'], y=comparison_df[f'{sample_student} Score'])
])

fig_student.update_layout(barmode='group', title=f'Performance of {sample_student} vs. Class Average by Subject')
fig_student.show()
```

**5. Supplementary Activity:**

Design Your Own Dashboard: “Smart Campus Energy Monitor”
Create a dashboard that shows the energy usage of different campus areas (e.g., laboratories, offices, classrooms) over time.

1. Simulate or use sample data (Energy consumption per hour or per day).

2. Include at least three charts (bar, line, pie).

3. Ensure your design reflects a clear purpose and target user (e.g., facilities manager).

4. Write a short explanation of your engineering decisions (why you chose certain colors, layouts, or interactivity).

```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

#Generate the weekly consumption
facilities = ["Classroom", "Registrar", "Offices", "Laboratories", "Canteen"]
days = pd.date_range("2025-10-27", periods=7, freq="D")

np.random.seed(42)
energy_data = {
    "Date": np.tile(days, len(facilities)),
    "Facility": np.repeat(facilities, len(days)),
    "Energy_Consumption_kWh": np.concatenate([
        np.random.normal(85, 5, 7),   # Classroom
        np.random.normal(45, 3, 7),   # Registrar
        np.random.normal(60, 4, 7),   # Offices
        np.random.normal(120, 8, 7),  # Laboratories
        np.random.normal(150, 10, 7)  # Canteen
    ])
}
energy_df = pd.DataFrame(energy_data)

# Generate daily usage hours
hours_data = {
    "Date": np.tile(days, len(facilities)),
    "Facility": np.repeat(facilities, len(days)),
    "Usage_Hours": np.concatenate([
        np.random.normal(8, 0.5, 7),  # Classroom
        np.random.normal(7, 0.5, 7),  # Registrar
        np.random.normal(9, 0.5, 7),  # Offices
        np.random.normal(10, 0.7, 7), # Laboratories
        np.random.normal(12, 1, 7)    # Canteen
    ])
}
hours_df = pd.DataFrame(hours_data)

# Create distribution of consumption across appliances to make the data realistic
appliances = {
    "Classroom": {"Lights": 30, "Fans": 25, "Projector": 20, "Sockets": 25},
    "Registrar": {"Lights": 35, "Aircon": 40, "Computers": 25},
    "Offices": {"Lights": 25, "Aircon": 45, "Computers": 30},
    "Laboratories": {"Lights": 20, "Computers": 30, "Equipment": 50},
    "Canteen": {"Lights": 15, "Fridge": 40, "Cooking_Appliances": 45}
}

appliance_data = []
for facility, items in appliances.items():
    for appliance, contribution in items.items():
        appliance_data.append({
            "Facility": facility,
            "Appliance": appliance,
            "Contribution_%": contribution
        })

appliance_df = pd.DataFrame(appliance_data)

#Obtain the key insights
avg_energy = energy_df.groupby("Facility")["Energy_Consumption_kWh"].mean().round(1)
avg_hours = hours_df.groupby("Facility")["Usage_Hours"].mean().round(1)
highlight = {
    "Highest Consumption": avg_energy.idxmax(),
    "Lowest Consumption": avg_energy.idxmin(),
    "Most Used Facility": avg_hours.idxmax(),
    "Least Used Facility": avg_hours.idxmin(),
}

#--Creating the Different Plots--#

# Pie Chart - Energy Consumption by Facility to provide general and foundational insights in energy
pie_fig = px.pie(
    values=avg_energy.values,
    names=avg_energy.index,
    title="Average Weekly Energy Consumption by Facility",
    hole=0.4
)
pie_fig.update_traces(textinfo='percent+label')

# Bar Chart - Average Daily Usage Hours to create a predictable expectation
bar_fig = px.bar(
    x=avg_hours.index,
    y=avg_hours.values,
    text=avg_hours.values,
    labels={"x": "Facility", "y": "Average Daily Usage Hours"},
    title="Average Daily Usage Hours per Facility"
)
bar_fig.update_traces(textposition="outside")

# Line Chart - This provides insights in contribution of appliances in energy consumption
line_fig = px.line(
    appliance_df,
    x="Appliance",
    y="Contribution_%",
    color="Facility",
    markers=True,
    title="Appliance Contribution to Energy Consumption"
)

#Combine to one dashboard
fig = make_subplots(
    rows=2, cols=2,
    specs=[[{"type": "domain"}, {"type": "xy"}],
           [{"type": "xy", "colspan": 2}, None]],
    subplot_titles=(
        "Energy Consumption Share",
        "Facility Usage Hours",
        "Appliance Contribution Trend"
    )
)

fig.add_trace(go.Pie(labels=avg_energy.index, values=avg_energy.values, hole=0.4), row=1, col=1)

fig.add_trace(go.Bar(x=avg_hours.index, y=avg_hours.values, text=avg_hours.values, textposition='outside'), row=1, col=2)

for facility in appliance_df["Facility"].unique():
    subset = appliance_df[appliance_df["Facility"] == facility]
    fig.add_trace(go.Scatter(x=subset["Appliance"], y=subset["Contribution_%"], mode='lines+markers', name=facility), row=2, col=1)

fig.update_layout(
    title_text="School Energy Consumption Dashboard for Facilities Engineers",
    height=800,
    showlegend=True
)

print("📊 Key Insights:")
for k, v in highlight.items():
    print(f"- {k}: {v}")

fig.show()
```

In this supplementary I designed the data dashboard to be specifically for facilities engineer. I followed the principles in choosing color for barchart by avoiding too colorful barcharts for unnecessary things, because on bar charts rankings are enough. Here I have added a pie chart for average consumption percentage which is an excellent start for deducing information from power. Then we follow up with the second chart containing the number of hours the facility is in use. At least we now have an idea, which facilities take longest usage time. This allows us to assess their usual times and also to determine in the future whether power consumption is unjustified on the number of hours the facility was used. Lastly I made a dataset for the common devices found in the different school facilities. This allows us to make sure that the appliances are monitored, and whether there could be a misuse evident with sudden spikes. Also for engineers maintaining the facility it is important to look whether some appliances consume more power, because it might signal a need for maintenance, fix, or replacement.
